**1. Breadth-First Search**

```python
# Membuat graph tree
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'I'],
    'D': ['I'],
    'E': [],
    'F': [],
    'G': [],
    'I': []
}
# Visit_complete = daftar yang digunakan untuk melacak simpul-simpul yang telah dikunjungi
# Grapgh = graph yang akan dijeajahi, biasanya dalam bentuk dictionary, dimana setiap simpul adalah kunci (key)
# dan tetangga-tetangga dari simpul tersebut disimpan dalam list sebagai milainya
# Current_node = simpul awal

def bfs(visit_complete, graph, current_node):

    # Menginisialisasi pencarian dengan menambahkan simpul awal ke dalam list visit-complete
    visit_complete.append(current_node)

    # Inisiasi queue
    queue = []

    # Simpul awal ditambahkan ke dalam queue
    queue.append(current_node)

    # Bila masih ada simpul yang harus dijelajahi
    while queue:

        # Mengambil simpul pertama dari antrian queue
        s = queue.pop(0) 

        # Simpul yang sedang dijelajahi
        print(s)

        # Mengiterasi melalui tetangga-tetangga dari simpul saat ini (s)
        for neighbour in graph[s]:

            # Cek apakah tetangga tersebut belum pernah dikunjungi
            if neighbour not in visit_complete:

                # Menambahkan tetangga tersebut ke list visit_complete
                visit_complete.append(neighbour)

                # Memasukkan ke queue
                queue.append(neighbour)

print("Hasil Algoritma Breadth-First Search: ")
# Panggil fungsi bfs
bfs([], graph, 'A')
```

**2. Contoh Lain Algoritma Breadth-First Search (BFS)**

```python
# Definisikan sebuah kasus graf
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': ['I'],
    'F': [],
    'G': ['J', 'K'],
    'H': [],
    'I': [],
    'J': [],
    'K': ['L'],
    'L': [],
}

def bfs(graph, start, end):
    # Buat queue untuk menyimpan simpul-simpul yang belum dikunjungi
    queue = []

    # Tambahkan simpul awal ke queue
    queue.append(start)

    # Buat list jalur
    path = []

    # Looping sampai queue kosong
    while queue:

        # Ambil simpul pertama dari queue
        node = queue.pop(0)

        # Jika simpul tersebut adalah simpul tujuan, maka retun jalur
        if node == end:
            return path + [node]
        
        # Tambahkan simpul tersebut ke jalur
        path.append(node)

        # Mengiterasi melalui tetangga-tetangga dari simpul saat ini
        for neighbor in graph[node]:

            # Cek apakah tetangga tersebut belum pernah dikunjungi
            if neighbor not in path:
            
                # Masukkan ke queue
                queue.append(neighbor)

# Kasus pencarian dari simpul A ke simpul L (akhir)
path = bfs(graph, 'A', 'L')

# Cetak jalut
print("Hasil Algoritma Breadth-First Search: ")
print(path)
```

**3. Depth-First Search (DFS)**

```python
# Algoritma DFS
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '2': [],
    '4': ['8'],
    '7': ['8'],
    '8': []
}

visited = set() # Set / himpunan graph yang dikunjungi

def dfs(visited, graph, node):
    if node not in visited:

        # Mencetak simpul saat ini yang sedang dijalani
        print(node)

        # Menandai simpul saat ini sebagai telah dikunjungi dan menaambahkan ke set valid
        visited.add(node)

        # Mengiterasi melalui tetangga-tetangga dari simpul saat ini
        for neighbour in graph[node]:

            # Memanggil fungsi dfs secara rekursif untuk menjelajahi tetangga yang belum pernah dikunjungi
            dfs(visited, graph, neighbour)

# Cetak jalur
print("Hasil Algoritma Depth-First Search: ")
dfs(visited, graph, '5')
```

**4. Contoh Lain dari Algoritma Depth-First Search (DFS)**

```python
# Definisikan graf berdasarkan gambar
graph = {
    '1': ['2', '5', '9'],
    '2': ['3'],
    '5': ['6', '8'],
    '9': ['10'],
    '3': ['4'],
    '6': ['7'],
    '8': [],
    '10': [],
    '4': [],
    '7': []
}

def dfs(graph, start, end):

    # Buat stack untuk menyimpan simpul-simpul yang belum dikunjungi
    stack = []

    # Tambahkan simpul awal ke stack
    stack.append(start)

    # Buat list jalur
    path = []
    
    # Looping sampai stack kosong
    while stack:

        # Ambil simpul pertama dari stack
        node = stack.pop()

        # Jika simpul tersebut adalah simpul tujuan, maka return jalur
        if node == end:
            return path + [node]
        # Tambahkan simpul tersebut ke jalur
        path.append(node)
        
        # Tambahkan semua simpul tetangga simpul tersebut ke stack, mulai dari yang paling kanan
        for neighbor in reversed(graph[node]):
            if neighbor not in path:
                stack.append(neighbor)

# Mulai pencarian, misal dari simpul "1" ke simpul "10"
path = dfs(graph, '1', '10')

# Cetak jalur
print("Hasil Algoritma Depth-First Search:")
print(path)
```

**5. Depth-Limited Search**

```python
# Definisikan graph
graph = {
    'S': ['A', 'B'],
    'A': ['C', 'D'],
    'B': ['I', 'J'],
    'C': ['E', 'F'],
    'D': ['G'],
    'I': ['H'],
    'J': [],
    'E': [],
    'F': [],
    'G': [],
    'H': [],
}

def dls(graph, start, end, depth):
    """
        graph: Graf yang akan dicari
        start: Simpul awal
        end: Simpul tujuan
        depth : Batas kedalaman pencarian
    """

    # Buat stack untuk menyimpan simpul-simpul yang belum dikunjungi
    stack = []

    # Tambahkan simpul awal ke stack
    stack.append((start, 0))

    # Buat list jalur
    path = []

    # Looping sampai stack kosong
    while stack:

        # Ambil simpul dan kedalamannya dari stack
        node, current_depth = stack.pop()

        # Jika simpul tersebut adalah simpul tujuan, maka return jalur
        if node == end:
            return path + [node]
        
        # Jika kedalaman pencarian belum mencapai batas, maka lanjutkan pencarian
        if current_depth < depth:
            # Tambahkan simpul tersebut ke jalur
            path.append(node)

            # Tambahkan semua simpul tetangga ke stack
            for neighbor in graph[node]:
                if neighbor not in path:
                    stack.append((neighbor, current_depth + 1))

# Mulai peencarian dari simpul "S" ke simpul "J" dengan kedalaman 2
path = dls(graph, "S", "J", 2)

# Cetak jalur
print("Hasil Algoritma Depth-Limited Search:")
print(path)
```

**6. Contoh Lain dari Algoritma Depth-Limited Search**

```python
# Definisikan graf
graph = {
    'A' : ['Z', 'S', 'T'],
    'Z' : ['O'],
    'S' : ['O', 'F', 'R'],
    'T' : ['L'],
    'O' : ['S'],
    'O' : [],
    'F' : ['B'],
    'R' : ['C', 'P'],
    'L' : ['M'],
    'S' : ['F', 'R'],
    'B' : [],
    'C' : [],
    'P' : [],
    'M' : ['D'],
    'F' : ['B'],
    'R' : [],
    'D' : [],
    'B' : []
}

def DLS(start, goal, path, level, maxD):
    """
    Args:
        start: Node awal pencarian.
        goal: Node tujuan yang ingin dicapai.
        path: Daftar jalur yang sedang dieksplorasi.
        level: Kedalaman saat ini.
        maxD: Kedalaman maksimum pencarian.
        visited: Set untuk menyimpan node yang sudah dikunjungi.
    """

    print(f"\nLevel sekarang: {level}")
    print(f"Memeriksa node: {start}")

    if start == goal:
        print("Tujuan ditemukan!")
        return path + [start]

    if level >= maxD:
        return None

    path.append(start)

    for neighbor in graph[start]:
            result = DLS(neighbor, goal, path.copy(), level + 1, maxD)
            if result:
                return result

    return None

# Contoh penggunaan
start_node = 'A'
goal_node = 'D'
max_depth = 3
path = []
result = DLS(start_node, goal_node, path, 0, max_depth)

if result:
    print("Jalur ke tujuan:", result)
else:
    print("Tujuan tidak ditemukan dalam batas kedalaman yang ditentukan.")
```

**7. Algoritma Uniform-Cost Search (UCS)**

```python
# Definisi graf
graph = {
    'S': {'A': 1, 'B': 4},
    'A': {'C': 3, 'D': 2},
    'B': {'G': 5},
    'C': {'E': 5},
    'D': {'G': 3},
    'E': {'G': 5},
    'G': {}
}

def ucs(graph, start, end):
    """
        graph: Graf yang akan dicari.
        start: Simpul awal.
        end: Simpul tujuan.
    """
    # Buat queue untuk menyimpan simpul-simpul yang belum di kunjungi
    queue = []

    # Tambahkan simpul awal ke queue dengan biaya awal 0
    queue = [(start, 0)] 

    # Buat dictionary untuk menyimpan jalur dan biaya setiap simpul
    path_cost = {start: 0}  
    path = {}  
    
    # Looping sampai queue kosong
    while queue:

        # Ambil simpul dan biayanya dari queue
        node, cost = queue.pop(0)  

        # Jika simpul tersebut adalah simpul tujuan, maka rekonstruksi jalur dan return
        if node == end: 
            path_nodes = []
            while node != start:
                path_nodes.append(node)
                node = path[node]
            path_nodes.append(start)
            return list(reversed(path_nodes))

        # Tambahkan semua simpul tetangga, simpul tersebut ke queue, diurutkan berdasarkan baiay
        for neighbor, neighbor_cost in graph.get(node, {}).items():
            new_cost = cost + neighbor_cost
            if neighbor not in path_cost or new_cost < path_cost[neighbor]:
                path_cost[neighbor] = new_cost
                queue.append((neighbor, new_cost))
                path[neighbor] = node

    # Jika tidak ada jalur yang ditemukan
    return None 

# Mulai pencarian dari simpul "S" ke simpul "G"
path = ucs(graph, "S", "G")

# Cetak jalur
if path:
    print("Jalur terpendek:", path)
else:
    print("Tidak ada jalur yang ditemukan.")
```

**8. Iterative-Deepening Dept First Search (IDDFS)**

```python
# Definisikan graf
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H', 'I'],
    'F': ['K'],
    'E': [],
    'G': [],
    'H': [],
    'I': [],
    'K': []
}

def iddfs(graph, start, end, depth):
    """
    graph: Graf yang akan dicari.
    start: Simpul awal.
    end: Simpul tujuan.
    depth: Batas kedalaman pencarian.
    """
    # Buat stack untuk menyimpan simpul-simpul yang belum dikunjungi
    stack = []
    
    # Tambahkan simpul awal ke stack
    stack.append((start, 0))
    
    # Buat list jalur
    path = []

    # Looping sampai stack kosong
    while stack:
        # Ambil simpul dan kedalamannya dari stack
        node, current_depth = stack.pop()

        # Jika simpul tersebut adalah simpul tujuan, maka return jalur
        if node == end:
            return path + [node]

        # Jika kedalaman pencarian belum mencapai batas, maka lanjutkan pencarian
        if current_depth < depth:
            # Tambahkan simpul tersebut ke jalur
            path.append(node)
            
            # Tambahkan semua simpul tetangga simpul tersebut ke stack, dengan kedalaman yang lebih besar
            for neighbor in graph[node]:
                if neighbor not in path:
                    stack.append((neighbor, current_depth + 1))
    
    # Jika tidak menemukan jalur
    return None

# Mulai pencarian dari simpul "A" ke simpul "G" dengan kedalaman 4
path = iddfs(graph, 'A', 'G', 4)

# Cetak jalur
print("Hasil Algoritma Iterative-Deepening Dept First Search:")
print(path)
```

**a. Buatlah program Python untuk menentukan jalur dimulai dari kota Frankfurt dengan mengimplementasikan algoritma BFS**

```python
from collections import deque

# Representasi graf dengan jarak yang tidak dihitung untuk BFS
graph = {
    'Frankfurt': ['Mannheim', 'Würzburg', 'Kassel'],
    'Mannheim': ['Karlsruhe'],
    'Würzburg': ['Nürnberg', 'Erfurt'],
    'Kassel': ['München'],
    'Karlsruhe': ['Augsburg'],
    'Nürnberg': ['Stuttgart', 'München'],
    'Erfurt': [],
    'Augsburg': ['München'],
    'Stuttgart': [],
    'München': []
}

def bfs(graph, start):
    # Queue untuk BFS
    queue = deque([start])
    visited = set([start])
    path = []

    while queue:
        city = queue.popleft()
        path.append(city)

        for neighbor in graph[city]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    
    return path

# Jalur dimulai dari Frankfurt
path = bfs(graph, 'Frankfurt')
print("Jalur BFS dari Frankfurt:")
print(path)
```

**b. Buatlah program Python untuk menentukan jalur dimulai dari kota Frankfurt dan berakhir pada kota Stuttgart dengan mengimplementasikan algoritma UCS**

```python
import heapq

# Definisikan graf dengan jarak antara kota-kota (berat setiap edge)
graph = {
    'Frankfurt': {'Mannheim': 85, 'Würzburg': 217, 'Kassel': 173},
    'Mannheim': {'Frankfurt': 85, 'Karlsruhe': 80},
    'Würzburg': {'Frankfurt': 217, 'Nürnberg': 103, 'Erfurt': 186},
    'Kassel': {'Frankfurt': 173, 'München': 502},
    'Karlsruhe': {'Mannheim': 80, 'Augsburg': 250},
    'Nürnberg': {'Würzburg': 103, 'Stuttgart': 183, 'München': 167},
    'Erfurt': {'Würzburg': 186},
    'Augsburg': {'Karlsruhe': 250, 'München': 84},
    'München': {'Kassel': 502, 'Nürnberg': 167, 'Augsburg': 84},
    'Stuttgart': {'Nürnberg': 183}
}

# Fungsi UCS untuk mencari jalur dengan biaya terendah
def ucs(graph, start, goal):
    queue = [(0, start, [])]  # (biaya, node saat ini, jalur)
    visited = set()  # Node yang sudah dikunjungi
    
    while queue:
        (cost, node, path) = heapq.heappop(queue)  # Ambil node dengan biaya terkecil
        if node not in visited:
            visited.add(node)
            path = path + [node]
            
            # Jika node tujuan tercapai, return jalur dan biayanya
            if node == goal:
                return (path, cost)
            
            # Periksa tetangga-tetangga node
            for neighbor, edge_cost in graph[node].items():
                if neighbor not in visited:
                    heapq.heappush(queue, (cost + edge_cost, neighbor, path))
    
    return None  # Jika tidak ada jalur ditemukan

# Cari jalur dari Frankfurt ke Stuttgart
result = ucs(graph, 'Frankfurt', 'Stuttgart')

# Cetak hasilnya
if result:
    path, cost = result
    print(f"Jalur terpendek dari Frankfurt ke Stuttgart: {path} dengan biaya {cost} km")
else:
    print("Jalur tidak ditemukan")
```
